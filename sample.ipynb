```python
import numpy
import matplotlib.pyplot as plt
import cv2 #opencv
```

```python
print(cv2.__version__)
```

```python
img = cv2.imread('images/photographer.jpg')
print(img.shape)
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
print(gray_img.shape)
```

```python
plt.figure(figsize=(8, 6), dpi=80)
plt.imshow(gray_img, cmap='gray', vmin=0, vmax=255)
```

```python
neg_img = 1 - img
plt.imshow(neg_img)
```

```python
cv2.imwrite('images/photographer_neg.jpg', neg_img)
```
